#### Plot the necessary charts to show effect of each feature on total_sales_price

```python
import matplotlib.pyplot as plt

# plot scatter plot for numerical columns
numerical_columns = ['carat_weight', 'depth_percent', 'table_percent', 'meas_length', 'meas_width', 'meas_depth','total_sales_price']
for col in numerical_columns:
    plt.scatter(df[col], df['total_sales_price'], alpha=0.2)
    plt.xlabel(col)
    plt.ylabel('total_sales_price')
    plt.show()
```

```python
df.skew()
```

```python
# plot bar plot for categorical columns
categorical_columns = ['cut', 'color', 'clarity', 'cut_quality', 'lab', 'symmetry', 'polish', 'eye_clean', 'culet_size', 'culet_condition', 'girdle_min', 'girdle_max', 'fluor_color', 'fluor_intensity', 'fancy_color_dominant_color', 'fancy_color_secondary_color', 'fancy_color_overtone', 'fancy_color_intensity']
for col in categorical_columns:
    grouped_df = df.groupby(col)['total_sales_price'].mean().reset_index()
    plt.bar(grouped_df[col], grouped_df['total_sales_price'])
    plt.xticks(rotation=90)
    plt.xlabel(col)
    plt.ylabel('total_sales_price')
    plt.show()
```
